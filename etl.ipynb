```python
import pandas as pd
```

```python
# load in csv
data_csv = "Resources/crime_data.xlsx"
perth_data_df = pd.read_excel(data_csv, sheet_name = 'Perth District')
```

```python
perth_data_df.head(15)
```

```python
dropped_rows = perth_data_df.iloc[7:,:]
dropped_rows
```

```python
dropped_columns = dropped_rows[['Region =', 'Unnamed: 30', 'Unnamed: 33', 'Unnamed: 35', 'Unnamed: 37', 'Unnamed: 39', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 47', 'Unnamed: 48']]
dropped_columns
```

```python
renamed_df = dropped_columns.rename(columns={
    "Region =" : "Month_and_Year",
    'Unnamed: 30' : 'Robbery',
    'Unnamed: 33' : 'Burglary', 
    'Unnamed: 35' : 'Stealing_of_Motor_Vehicle', 
    'Unnamed: 37' : 'Stealing_From_Motor_Vehicle', 
    'Unnamed: 39' : 'Stealing_From_Dwelling', 
    'Unnamed: 44' : 'Criminal_Damage_to_Property', 
    'Unnamed: 45' : 'Damage_to_Property', 
    'Unnamed: 47' : 'Cause_Damage_by_Fire', 
    'Unnamed: 48' : 'Damage_from_Bushfire', 
   })
renamed_df.head()
```

```python
renamed_df.to_csv("Resources/perth.csv", encoding="utf-8", index=False, header=True)
```
